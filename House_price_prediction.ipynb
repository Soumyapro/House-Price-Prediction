```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
sns.boxplot(data=new_df,orient='h')
plt.title("Box Plot for all features")
plt.xlabel("Values")
plt.show()
```

```python
corr = new_df.corr()
plt.figure(figsize=(14, 10))
cmap = sns.color_palette("Reds", as_cmap=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap=cmap,linewidths=2)
plt.title('Correlation Heatmap')
plt.yticks(rotation=0)
plt.show()
```

```python
sns.set(style="whitegrid", context="notebook")
plt.figure(figsize=(12, 8))

for col in new_df.columns:
    sns.kdeplot(new_df[col], label=col)

plt.title('KDE Plots for Different Variables')
plt.legend()
plt.show()
```

```python
sns.set(style="whitegrid", context="notebook")
plt.figure(figsize=(12, 8))
sns.kdeplot(new_df['medv'],label=new_df['medv'])
plt.title('KDE Plot for MEDV')
plt.legend()
plt.show()
```

```python
x1 = new_df[['nox','rm','dis','ptratio','lstat']]
y1 = new_df['medv']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size=0.33, random_state=5)
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
dff = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
dff.head(20)
```

```python
from sklearn.svm import SVR

svr = SVR(kernel='linear')
svr.fit(x_train,y_train)
y_pred_svm = svr.predict(x_test)
r2_svm = r2_score(y_test,y_pred_svm)
print(r2_svm)
```

```python
from sklearn import metrics
mean_absolute_error = metrics.mean_absolute_error(y_test, y_pred)
mean_squared_error = metrics.mean_squared_error(y_test, y_pred)
root_mean_squared_error = np.sqrt(mean_squared_error)
r2_score = metrics.r2_score(y_test, y_pred)
print("Mean Absolute Error : ", mean_absolute_error)
print("Mean Squared Error : ",mean_squared_error)
print("Root Mean Squared Error : ",root_mean_squared_error)
print("R2 Score : ",r2_score)
```

```python
x2 = np.array(x_test['nox'])
y2 = np.array(y_pred)
m, b = np.polyfit(x2, y2, 1)

plt.plot(x_test['nox'], m*x_test['nox'] + b, label='Regression Line')
plt.plot(x_test['nox'], y_pred, 'o', label='Data Points')
plt.xlabel("Nitric Oxide Content")
plt.ylabel("Median Values")
plt.legend()
plt.show()
```
